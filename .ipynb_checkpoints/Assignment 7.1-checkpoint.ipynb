## Data Cleaning and Preparation Process for Canada Immigration Dataset

This section describes the steps followed to clean, standardize, and prepare the Canada
immigration dataset for analysis. The objective was to retain meaningful descriptive
attributes, ensure consistency across columns, and compute a reliable total immigration
measure for each country.

---

```python
import pandas as pd
```

```python
!pip install openpyxl
```

```python
# read the data into a pandas dataframe
df = pd.read_excel('Immigration.xlsx', skiprows=range(20), skipfooter=2)
df
```

```python
# dimensions of the data
df.shape
```

### 1. Removal of Unnecessary Columns

The raw dataset contained several administrative and metadata columns that were not
required for analytical purposes. These columns did not contribute to country-level
immigration analysis and were therefore removed to simplify the dataset.

**Columns removed:**
- Type  
- Coverage  
- AREA  
- REG  
- DEV  

This step helped reduce noise and improved the readability of the dataset.

---

```python
# remove the unnecessary columns
df.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis=1, inplace=True)
df.head()
```

### 2. Column Renaming for Clarity

To improve interpretability, selected columns were renamed using intuitive and
self-explanatory labels. This ensured that column names clearly reflected their contents.

**Column mapping:**
- `OdName` → `Country`  
- `AreaName` → `Continent`  
- `RegName` → `Region`  
- `DevName` → `DevName` (retained as-is for development classification)

---

```python
# rename the columns for simplicity
df.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)
df.head()
```

### 3. Standardization of Column Labels

The year columns in the dataset were originally stored as numeric (float) values.
For consistency and to avoid issues during column selection and plotting, all column
labels were converted to string type.

This standardization ensured uniform handling of column names throughout the analysis.

---

```python
df.columns = df.columns.astype(str)
```


### 4. Definition of Year Columns

A list of year columns ranging from **1980 to 2013** was explicitly defined.
This step ensured that only valid year-based immigration data would be included in
subsequent aggregation operations.

Explicit year selection prevents accidental inclusion of non-year or derived columns.

---

```python
year_columns = [str(year) for year in range(1980, 2014)]
df.head()
```

### 5. Computation of Total Immigration per Country

A new column named **`Total`** was added to the dataset. This column represents the
cumulative number of immigrants arriving in Canada from each country across all years
from 1980 to 2013.

Only the year columns were included in the calculation to avoid double counting or
type-related errors. This approach ensured accurate aggregation and consistency with
official reference values.

---

```python
df['Total'] = df[year_columns].sum(axis=1)
```

```python
df.shape
```

### 6. Dataset Verification

Finally, the structure of the cleaned dataset was verified by checking its dimensions
and previewing the first few rows. This confirmed that:

- All required descriptive columns were retained
- Year columns were correctly formatted
- The `Total` column accurately reflected cumulative immigration values

The resulting dataset is clean, well-structured, and ready for analysis and visualization.
