## Objective  
The objective of this activity is to apply the K-Means clustering algorithm to the **Sales Transaction Dataset Weekly** in order to identify products with similar sales trends. Since the dataset contains only sales data and no product-related information, clustering is performed purely based on sales patterns.

---

## Dataset Description  
The **Sales Transaction Dataset Weekly** contains weekly sales data for **800 products over a period of one year**.  
Each row represents a product, and each column represents the weekly sales quantity.  
The dataset does not include any additional product information such as category, price, or brand.

---

## Problem Statement  
The dataset does not provide predefined labels or product attributes. Therefore, an **unsupervised learning** approach is required to discover hidden patterns in the data. The goal is to group products that show similar weekly sales behavior using the K-Means clustering algorithm.

---

```python
import pandas as pd
```

```python
file_url = 'https://raw.githubusercontent.com/PacktWorkshops/The-Applied-Artificial-Intelligence-Workshop/master/Datasets/Sales_Transactions_Dataset_Weekly.csv'
df = pd.read_csv(file_url)
df
```

- Dataset contains **811 rows** (each row = 1 product) and **107 columns**.  
- Column breakdown:  
  - **Product_Code** (first column)  
  - **52 weekly sales columns** starting with `W` (W0 to W51)  
  - **52 normalized sales columns** (better for K-Means clustering)  
  - **MIN and MAX columns** (not needed for clustering)  
- **Recommendation for K-Means:**  
  - Use **normalized columns** instead of absolute weekly sales  
  - **Remove**:  
    - `Product_Code` column  
    - All `W` columns  
    - `MIN` and `MAX` columns  
- Note: Weeks are numbered **0 to 51**, not 1 to 52.


- Keep only the **first 55 columns** (normalized sales).  
- Remove unnecessary columns using **`inplace=False`**.  
- Dataset now ready for **K-Means clustering**.


```python
df2 = df.drop(df.iloc[:,0:55], inplace = False,axis =1)
df2.head()
```

- Import KMeans from `sklearn.cluster`.  
- Create a **K-Means model** with:
  - `n_clusters = 8`  
  - `random_state = 8`  
- Fit the model on your dataset (`df2`).  
- Access:
  - **Cluster labels** → `kmeans_model.labels_`  
  - **Cluster centers** → `kmeans_model.cluster_centers_`


```python
from sklearn.cluster import KMeans
k_means_model = KMeans(n_clusters = 8 , random_state = 8)
k_means_model.fit(df2)
labels = k_means_model.labels_
labels
```

### Remove all unnecessary columns and add labels in df

```python
df.drop(df.iloc[:,53:],inplace =True, axis = 1)
df.drop('Product_Code',inplace = True, axis = 1)
df['labels'] = labels
df.head()
```

##  yearly average sale of each cluster

- **Group by cluster** and sum each product's weekly sales → `df_agg`.  
- **Count total products** per cluster using `count()` on a single column.  
- **Sum all sales columns** in `df_agg` → total sales per cluster.  
- **Calculate yearly average sales**: `total_sales / count_product`.  
- **Sort clusters** by highest yearly average sales → `df_final`.


```python
# Aggregate total weekly sales per cluster
df_agg = df.groupby('labels').sum()

# Count products per cluster
df_final = df.groupby('labels')['W0'].count().rename('count_product').to_frame()

# Calculate total and yearly average sales
df_final['total_sales'] = df_agg.sum(axis=1)
df_final['yearly_average_sales'] = df_final['total_sales'] / df_final['count_product']

# Sort clusters by yearly average sales
df_final = df_final.sort_values(by='yearly_average_sales', ascending=False)

df_final
```

```python
import matplotlib.pyplot as plt


# Reset index to make 'label' a column for plotting
df_plot = df_final.reset_index()

# Sort by yearly_average_sales (highest to lowest)
df_plot = df_plot.sort_values(by='yearly_average_sales', ascending=False)

# Bar plot for yearly average sales per cluster
plt.figure(figsize=(10,6))
bars = plt.bar(df_plot['labels'].astype(str), df_plot['yearly_average_sales'], color='skyblue')

# Add values on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 20, f"{yval:.2f}", ha='center', fontsize=10)

# Titles and labels
plt.title("Yearly Average Sales per Cluster", fontsize=16)
plt.xlabel("Cluster Label", fontsize=12)
plt.ylabel("Yearly Average Sales", fontsize=12)
plt.ylim(0, df_plot['yearly_average_sales'].max() * 1.1)  # extra space on top

plt.show()

```
